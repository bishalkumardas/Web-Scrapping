```python
#importing important libraries
import requests
from bs4 import BeautifulSoup
import pandas as pd
from tqdm import tqdm
```

```python

url="https://www.ambitionbox.com/list-of-companies?page=1"
headers = {
    'User-Agent': 'Mozilla/5.0 (Macintosh; Intel Mac OS X 10.12; rv:55.0) Gecko/20100101 Firefox/55.0',
}
```

```python
r=requests.get(url, headers=headers)
page=r.content
soup = BeautifulSoup(page,'html.parser')
```

```python
cname = soup.findAll(attrs = {"class":"company-name bold-title-l"})
```

```python
rating=soup.findAll(attrs = {"class":"rating-wrapper"})
```

```python
review=soup.findAll(attrs = {"class":"review-count sbold-Labels"})
```

```python
company_type=soup.findAll(attrs = {"class":"infoEntity sbold-list-header"})
```

```python
head_quarters=soup.findAll(attrs = {"class":"company-basic-info"})
```

```python
age=soup.findAll(attrs = {"class":"company-basic-info"})
```

```python
n_employee=soup.findAll(attrs = {"class":"company-basic-info"})
```

```python
age[1].text.split(' ')#.split('\t')[6]#.split('\n')[0]
```

```python
review[0].text.split('\n')[1].split('\t')[9].split('(')[1].split(')')[0]
```

```python
name=[]
rating1=[]
review1=[]
type=[]
head_quarters1=[]
age_1=[]
n_employee1=[]

for i in range(len(cname)):
    name.append(cname[i].text.split('\t')[9].split('\n')[0])
    rating1.append(rating[i].text.split('\n')[1].split('\t')[-1])
    review1.append(review[i].text.split('\n')[1].split('\t')[9].split('(')[1].split(')')[0])
    type.append(company_type[i].text.split('\t')[7].split('\n')[0])
    head_quarters1.append(head_quarters[i].text.split(' ')[1].split('\t')[7].split('\n')[0])
    age_1.append(age[i].text.split('\n')[5].split('\t')[7].split(' ')[0])
    n_employee1.append(n_employee[i].text.split('\n')[7].split('\t')[7])


```

```python
df_data={'Name':name,'Rating':rating1,'Review':review1,'Company Type':type,'Head Quarter':head_quarters1,'Company Age(in years)':age_1,'Number of Employee':n_employee1}
```

```python
df=pd.DataFrame(df_data)
```
