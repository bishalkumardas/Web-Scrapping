```python
#importing important libraries
import requests
from bs4 import BeautifulSoup
import pandas as pd
from tqdm import tqdm
```

```python
#importing and creating url dataframe 
data=pd.read_csv('Input.csv', nrows=150)
url="https://insights.blackcoffer.com/is-telehealth-the-future-of-healthcare-3/"
headers = {
    'User-Agent': 'Mozilla/5.0 (Macintosh; Intel Mac OS X 10.12; rv:55.0) Gecko/20100101 Firefox/55.0',
}
df = pd.DataFrame(data,columns=['URL','Title','Content'])
df


```

```python
#Scrapping the data from urls
url_list=[]
title_list=[]
content_list=[]
for i in tqdm(data['URL']):
        r=requests.get(i, headers=headers)
        page=r.content
        soup = BeautifulSoup(page,'html.parser')
        title = soup.findAll(attrs = {'class':'td-crumb-container'})
        title = title[0].text.split('  ')[2]
        content = soup.find(attrs = {'td-post-content'}).text.replace('\n',' ')
        data = [[i,title,content]]
        title_list.append(title)
        url_list.append(i)
        content_list.append(content)
```

```python
#storing the data in a dataframe
df_data={'Url':url_list,'Title':title_list,'content':content_list}
df=pd.DataFrame(df_data)
```
